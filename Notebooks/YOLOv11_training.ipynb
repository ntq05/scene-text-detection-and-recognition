# 1. Import Libraries

```python
from ultralytics import YOLO
```

# 2. Training

```python
yolo_yaml_path = "/content/Datasets/yolo_data/data.yaml"
```

```python
model_1 = YOLO("yolo11m.pt")
```

```python
model_2 = YOLO("yolo11n.pt")
```

```python
results_1 = model_1.train(
    data=yolo_yaml_path,
    epochs=100,
    imgsz=640,
    cache=True,
    patience=20,
    plots=True,
    seed=42,
    name="yolo11m_seed42"
)
```

```python
results_2 = model_2.train(
    data=yolo_yaml_path,
    epochs=100,
    imgsz=640,
    cache=True,
    patience=20,
    plots=True,
    seed=42,
    name="yolo11n_seed42"
)
```

# 3. Evaluation

```python
model_1_path = "/content/runs/detect/yolo11m_seed42/weights/best.pt"
model_2_path = "/content/runs/detect/yolo11n_seed42/weights/best.pt"
```

```python
model_1 = YOLO(model_1_path)

metrics_1 = model_1.val(
    data="Datasets/yolo_data/data.yaml",
    split="val"
)
```

```python
metrics_1 = model_1.val(
    data="Datasets/yolo_data/data.yaml",
    split="test"
)
```

```python
model_2 = YOLO(model_2_path)

metrics_2 = model_2.val(
    data="Datasets/yolo_data/data.yaml",
    split="val"
)
```

```python
metrics_2 = model_2.val(
    data="Datasets/yolo_data/data.yaml",
    split="test"
)
```
